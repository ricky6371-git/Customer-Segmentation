```python
features = ['Age','Income', 'Total_spending','NumWebPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Recency']
```

```python
X = df[features].copy()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2, 10), wcss, marker='o')
plt.title("Elbow Method for Optimal k") 
plt.xlabel("Number of clusters (k)")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=6, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
cluster_labels = {
    0: "High income, high spending â†’ Premium Customers",
    1: "Medium income, high spending â†’ Value Seekers",
    2: "High web purchases, low store purchases â†’ Digital Buyers",
    3: "Low income, medium spending â†’ Thrifty Customers",
    4: "Young, medium income â†’ Potential Loyalists",
    5: "Low recency, inactive â†’ Dormant Customers",
    6: "Low income, low spending â†’ Budget Customers"
}

df['Cluster_Label'] = df['Cluster'].map(cluster_labels)
```

```python
df
```

```python
cluster_summary = df.groupby('Cluster')[features].mean()
```

```python
cluster_summary
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(X_scaled) 
df["PCA1"] = pca_data[:, 0]
df["PCA2"] = pca_data[:, 1]
```

```python
pca_data
```

```python
sns.scatterplot(data=df, x="PCA1", y="PCA2", hue="Cluster", palette="Set1")
plt.title("Customer Segments based on PCA")
plt.show()
```
